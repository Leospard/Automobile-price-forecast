```python
import requests
import pandas as pd
import html
from lxml import etree
import re
import time
import csv
```

```python
def get_html(brand, page):
#     url = f'https://www.che168.com/china/a0_0msdgscncgpi1ltocsp{page}exx0/'
    url = f'https://www.che168.com/china/{brand}/a0_0msdgscncgpi1ltocsp{page}exx0/?pvareaid=102179#currengpostion'
    r = requests.get(url)
    r = html.unescape(r.text)
    return r
```

```python
for i in range(50,101):
    while 1:
        raw_html = get_html(i)
        print("正在爬取第" + str(i) + "页")
        t_html = etree.HTML(raw_html)
        divs = t_html.xpath('.//div[@class="tp-cards-tofu fn-clear"]')
        if len(divs) == 0:
            time.sleep(5)
        else:
            break
    #
    milage = divs[0].xpath("//li/@milage")
    price = divs[0].xpath("//li/@price")
    regdate = divs[0].xpath("//li/@regdate")
    name = divs[0].xpath('//h4[@class="card-name"]/text()')
    unit = divs[0].xpath('//p[@class="cards-unit"]/text()')
    region = [unit[i].split('／')[-2] for i in range(len(unit))]
    
    if len(milage) == len(price) == len(name) == len(regdate) == len(region):
        for j in range(len(name)):
            with open('YYY.csv', 'a+', newline='') as f:
                csv_write = csv.writer(f)
                data_row = [name[j], regdate[j], region[j], milage[j], price[j]]
                csv_write.writerow(data_row)
            f.close()
    else:
        print("Data not aligned!")
        break;
    time.sleep(2)

    
```
