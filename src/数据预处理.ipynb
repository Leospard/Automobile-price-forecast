#### 计算车龄CarAge

```python
month_list = {"Jan": 1, "Feb": 2, "Mar": 3, "Apr": 4, "May": 5, "Jun": 6, "Jul": 7, "Aug": 8, "Sep": 9, "Oct": 10, "Nov": 11, "Dec": 12}
```

```python
import re
def calMonths(dt, year = 22, month = 1):
    """
    根据三种格式的数据计算车龄（月）
    """
    if (dt == "未上牌"):
        return 0
    temp = dt.split('-')
    if (re.match(r".*[a-zA-Z]$", dt)):
        return 12*(year - int(temp[0])) + (month - month_list[temp[1]])
    elif (re.match(r"^[a-zA-Z]", dt)):
        return 12*(year - int(temp[1])) + (month - month_list[temp[0]])
    else:
        return 12*(2022 - int(temp[0])) + (month - int(temp[1]))
```

```python
data["CarAge"] = data["Date"].apply(calMonths)
data
```

#### 车龄标准化

```python
from sklearn.preprocessing import StandardScaler
scaler_Miles = StandardScaler()
data['CarAge_std'] = scaler_Miles.fit_transform(data['CarAge'].values.reshape(-1, 1))
data
```
