#### 计算车龄CarAge

```python
month_list = {"Jan": 1, "Feb": 2, "Mar": 3, "Apr": 4, "May": 5, "Jun": 6, "Jul": 7, "Aug": 8, "Sep": 9, "Oct": 10, "Nov": 11, "Dec": 12}
```

```python
import re
def calMonths(dt, year = 22, month = 1):
    """
    根据三种格式的数据计算车龄（月）
    """
    if (dt == "未上牌"):
        return 0
    temp = dt.split('-')
    if (re.match(r".*[a-zA-Z]$", dt)):
        return 12*(year - int(temp[0])) + (month - month_list[temp[1]])
    elif (re.match(r"^[a-zA-Z]", dt)):
        return 12*(year - int(temp[1])) + (month - month_list[temp[0]])
    else:
        return 12*(2022 - int(temp[0])) + (month - int(temp[1]))
```

```python
data["CarAge"] = data["Date"].apply(calMonths)
data
```

#### 车龄标准化

```python
from sklearn.preprocessing import StandardScaler
scaler_CarAge = StandardScaler()
data['CarAge_std'] = scaler_CarAge.fit_transform(data['CarAge'].values.reshape(-1, 1))
data
```

不同能源类型的样本数量相差较大，而其中汽油、柴油、汽油+48V清混系统、汽油+90V清混系统都属于传统燃油车

#### 合并近似类型

```python
def getEnergyType(EnergyType):
    if (EnergyType.startswith("汽油") or EnergyType == "柴油"):
        return "ICE" # 燃油车
    elif (EnergyType == "油电混合" or EnergyType == "插电式混合动力" or EnergyType == "增程式"):
        return "EV" # 新能源车
    else:
        return None
```

```python
data["EnergyType"] = data['能源类型'].apply(getEnergyType)
data
```

#### 编码

```python
from sklearn.preprocessing import LabelEncoder
le_EnergyType = LabelEncoder()
data['EnergyType_label'] = le_EnergyType.fit_transform(data['EnergyType'])
# 通过le.inverse_transform([1])可以获得指定编码对应的车名，即逆编码
# le_Level.classes_查看编码的原始信息
data
```

### 发动机

#### 查看

```python
data["发动机"]
```

发动机参数分为排量（是否涡轮增压），马力（功率），发动机气缸数量和布置形式

#### 分列

```python
# 获取排量
def getDisplacement(s):
    s = s["发动机"]
    if (s.startswith("增程式")):
        return None, None, None
    else:
        s = s.split()
    if (s[0][-1] == "T"):
        return float(s[0][:-1]), True, s[2]
    elif (s[0][-1] == "L"):
        return float(s[0][:-1]), False, s[2]
    else:
        return float(s[0][:-1]), None, s[2]
```

```python
data[["Displacement", "Turbo", "Engine"]] = data.apply(getDisplacement, axis=1, result_type="expand")
```

```python
# 排量
data["Displacement"].describe()
```

```python
# 是否涡轮增压
data["Turbo"].value_counts()
```

```python
# 发动机布置类型
data["Engine"].value_counts()
```
